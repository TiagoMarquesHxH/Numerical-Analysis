### Métodos de Análise numérica - Método de Euler

Este método de Análise numérica é rela

```python
import matplotlib.pyplot as plt

l = 0.1

def malthus(x):
    m = l*x
    return m


def euler(x,h):
    """ Função que aplica o método de Euler na equação de Malthus
    
    Args:
        x = valor inicial
        h = passos dados
        
    Returns:
        Uma lista contendo a aproximação feita pelo método de Euler para a equação de Malthus, e outra lista contendo a quantidade de iterações feitas para aproximar
    """

    eixo_x = []
    eixo_y = []
    v = (i * h for i in range(0, 20))
    for i in v:
        x1 = x + (h*malthus(x))
        eixo_x.append(x1)
        eixo_y.append(i)
        x = x1
        print(x1)
    return eixo_x, eixo_y
```

```python
print("Para h = 1")
print()

plt.title("EDO Malthus pelo método de Euler (h = 1)")
plt.xlabel('Iterações')
plt.ylabel('Aproximação')
plt.plot(euler(0.1,1)[1], euler(0.1,1)[0], c = 'b')
print(euler(0.1,1))
```
